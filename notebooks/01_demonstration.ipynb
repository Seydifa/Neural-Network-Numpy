# Neural Network from Scratch with NumPy

This notebook demonstrates the capabilities of our custom neural network implementation.

```python
import numpy as np
import plotly.graph_objects as go
from neural_network_numpy import NeuralNetwork
from neural_network_numpy.activations import sigmoid, dsigmoid, tanh, dtanh, linear, dlinear, relu, drelu
from neural_network_numpy.losses import mse, dmse, log_loss, dlog_loss
from neural_network_numpy.utils import plot_decision_boundary, to_categorical
```

## 1. Regression Example: Approximating a Function

We will approximate the function $f(x) = \cos(2x) + x\sin(3x) + \sqrt{x} - 2$.

```python
def f(x):
    return np.cos(2*x) + x*np.sin(3*x) + x**0.5 - 2

X = np.linspace(0, 5, 100).reshape(-1, 1)
Y = f(X)

model = NeuralNetwork(input_dim=1)
model.add_layer(10, activation=tanh, dactivation=dtanh)
model.add_layer(10, activation=tanh, dactivation=dtanh)
model.add_layer(1, activation=linear, dactivation=dlinear)

model.fit(X, Y, epochs=1000, lr=0.01, loss_fn=mse, dloss_fn=dmse, verbose=0)

Y_pred = model.forward(X)

fig = go.Figure()
fig.add_trace(go.Scatter(x=X.ravel(), y=Y.ravel(), name='Ground Truth'))
fig.add_trace(go.Scatter(x=X.ravel(), y=Y_pred.ravel(), name='Prediction'))
fig.update_layout(title='Function Approximation', template='plotly_white')
fig.show()
```

## 2. Classification Example: Moons Dataset

```python
from sklearn.datasets import make_moons
X_moons, y_moons = make_moons(n_samples=200, noise=0.1, random_state=42)
y_moons_cat = y_moons.reshape(-1, 1)

moon_model = NeuralNetwork(input_dim=2)
moon_model.add_layer(8, activation=sigmoid, dactivation=dsigmoid)
moon_model.add_layer(1, activation=sigmoid, dactivation=dsigmoid)

moon_model.fit(X_moons, y_moons_cat, epochs=1000, lr=0.1, loss_fn=log_loss, dloss_fn=dlog_loss, verbose=0)

plot_decision_boundary(moon_model, X_moons, y_moons)
```
